```python
import pandas as pd 
import numpy as np
from pprint import pprint
from collections import Counter
import uuid
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,classification_report
```

```python
# load
df = pd.read_csv('heart.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
# remove records with missing heart rate values
# df = df[df['max heart rate'].notnull()]
# df = df[df['oldpeak']>=0]

df['oldpeak'] = df['oldpeak'].replace(-99.99, np.nan)
df = df.fillna(df.median())
```

```python
# create testing and training sets
x = df.loc[:, df.columns != 'target']
y = df['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=2)


# lst = ['age', 'resting blood pressure', 'chol', 'max heart rate']
# x_train[lst] = StandardScaler().fit_transform(x_train[lst])
# x_test[lst] = StandardScaler().fit_transform(x_test[lst])

sc = MinMaxScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

x_train = pd.DataFrame(x_train,columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach','exang', 'oldpeak', 'slope', 'ca', 'thal'])
x_test = pd.DataFrame(x_test,columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach','exang', 'oldpeak', 'slope', 'ca', 'thal'])
```

```python
clf = XGBClassifier(learning_rate=0.1)
clf.fit(x_train, y_train)
pred = clf.predict(x_test)
conf_matrix = confusion_matrix(y_test, pred)
acc_score = accuracy_score(y_test, pred)
print(conf_matrix)
print(acc_score)
print(classification_report(y_test,pred))
```
