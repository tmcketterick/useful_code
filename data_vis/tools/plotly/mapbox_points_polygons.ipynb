```python
import json
import colorsys
import numpy as np
import pandas as pd
import geopandas as gpd
import plotly.graph_objects as go
```

```python
# load example data
gdf = gpd.read_file(gpd.datasets.get_path('nybb'))

# reproject (for geojson suitability)
gdf = gdf.to_crs({'init': 'epsg:4326'})

# convert to geojson and extract list of poylgons
geo_json_polygons = []
jsn = json.loads(gdf.to_json())
for feature in jsn['features']: 
        geo_json_polygons.append({"type": "FeatureCollection", 'features': [feature]})
num_polygons = len(geo_json_polygons)   

# define lat-lon points (using polygon centroids)
centroid_lat = gdf.centroid.y.unique()
centroid_lon = gdf.centroid.x.unique()
```
